# Exploratory Data Analysis (EDA)

This notebook provides a basic exploration of the sales dataset included in this repo.

```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plots
%matplotlib inline
sns.set(style='whitegrid')
```

## Load Dataset

We will use the existing `sample_sales.csv` in the `data/` folder.

```python
# Load CSV dataset
df = pd.read_csv('../data/sample_sales.csv')
df.head()
```

## Basic Information

```python
# Dataset shape and info
print("Shape:", df.shape)
print("\nInfo:")
print(df.info())

# Summary statistics
df.describe(include='all')
```

## Check for Missing Values

```python
df.isnull().sum()
```

## Simple Visualizations

```python
# Histogram of numerical columns
df.select_dtypes(include=['int64','float64']).hist(figsize=(12,8), bins=20, color='skyblue')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap (numerical columns)
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
